# if-else statements

This offers the next step in an if statement... what if the condition fails?
We have if-else.

### if-elif-else: are been used to automate the decision making process

```python
## if-else are like switches, binary operation, boolean values 
age = 54
if age >=20: 
    ## fails 
    print("You are allowed to join")
else: 
    ### this will execute 
    print("Sorry, you cannot join")
```

# if-elif-else

Let's keep building on what we have learned so far. 

BTW. elif = else if.

#### if condition1:
     action 1    
#### elif condtion2:
     action 2 
#### elif condtion3:
     action 3
#### else: 
      action 4

```python
age = 19 ### user defined age 
if age >=20:
    print("You are allowed to join")
elif age == 18:
    print("You can join for an extra $100")
elif age == 19:
    print("You can join for an extra $50")
else:
    print("Sorry, you cannot join")
```

```python
# You can use more than one elif
age = 17
if age >=20:
    print("You are allowed to join")
elif age == 18:
    print("You can join for an extra $100")
elif age == 17:
    print("You can join for an extra $125")
else:
    print("Sorry, you cannot join")
```
